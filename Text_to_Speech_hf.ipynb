```python
from transformers import pipeline
```

```python
pipe = pipeline('text-to-speech',model = 'suno/bark',device='cuda')
```

```python
text = "hi my name is jaskirat ad i am doing my bachelor's in artificial intelligence and data science"
out = pipe(text)
```
